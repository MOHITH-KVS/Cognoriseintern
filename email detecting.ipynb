Required libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Collecting and preprocess the data


```python
#loading the data from csv file to dataframe of panda
mail_data = pd.read_csv('/content/spam.csv')
```

```python
print(mail_data)
```

```python
#if any null values, there are replacing with a null string
check_data = mail_data.where((pd.notnull(mail_data)),'')
```

```python
print(check_data)
```

```python
#checking the size of the data set (rows and columns)
check_data.shape
```

Label encoding
Assuming that spam mail as 0 and ham mails as 1

```python
#spam --> as 0 and ham --> as 1
check_data.loc[check_data['Category']== 'spam', 'Category',] = 0
check_data.loc[check_data['Category']== 'ham', 'Category',]= 1
```

separating the data into two parts

```python
X = check_data['Message']
Y = check_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Now splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.2, random_state=3)
```

check how many rows  are splitted

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature extraction


```python
#transform the text data to numerical values (feature vectors) these values helps to acts a inputs to the logistic regression
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english',lowercase= True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

#coverting the y_train and y_test

Y_train= Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

Training the data to the model (logistic regression)

```python
model = LogisticRegression()
```

Now train the model with training data

```python
model.fit(X_train_features,Y_train)
```

Evaluating the trained model

```python
#prediciton on the trained data
prediction_on_trainedData = model.predict(X_train_features)
accuracyscore_on_trainedData = accuracy_score(Y_train,prediction_on_trainedData)
```

```python
print(accuracyscore_on_trainedData)
```

```python
#prediciton on the tested data
prediction_on_testedData = model.predict(X_test_features)
accuracyscore_on_testedData = accuracy_score(Y_test,prediction_on_testedData)
```

```python
print(accuracyscore_on_testedData)
```

Final part (Prediciting system)

```python
User_input = ["Nah I don't think he goes to usf, he lives around here though"]
User_input_extraction = feature_extraction.transform(User_input)
prediciton = model.predict(User_input_extraction)
print(prediciton)
```

```python
if prediciton[0] == 1:
  print("Given mail is 'Ham mail'")
else:
  print("Given mail is 'Spam mail'")
```
