Required libraries

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset from sklearn
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
 #loading the data into dataframe
 data_frame = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
```

```python
print(data_frame)
```

```python
data_frame.head()
```

```python
#adding the target value to the dataset (dataframe)
data_frame['diagnosis'] = breast_cancer.target
```

```python
print(data_frame)
```

```python
data_frame.head()
```

```python
data_frame.info()
```

```python
data_frame.shape
```

```python
#checking for missing values in the datasets
data_frame.isnull().sum()
```

from the above 1 --> benign and 0 --> malignant so there are 357 benign and 212 malignant

```python
#find the statistical values for target variabels which are 0 and 1
data_frame.groupby('diagnosis').mean()
```

Separting the data into target and featuresd which consider as X and Y

```python
X = data_frame.drop(columns = 'diagnosis', axis=1)
Y = data_frame['diagnosis']
```

```python
X.shape
```

```python
Y.shape
```

```python
#Now splitting the data into two parts which are train and test
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
#model training and implementing
model = LogisticRegression()
```

```python
#train the model by giving traing data
model.fit(X_train, Y_train)
```

```python
#Accuracy score on training data
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print(training_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
print(testing_accuracy)
```
